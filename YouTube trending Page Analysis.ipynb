# Importing Libraries 

```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data 

```python
youtube_india_data=pd.read_csv('IN_youtube_trending_data.csv')
```

# Structure of data

```python
youtube_india_data.info()
youtube_india_data
```

# Cleaning of Data

```python
youtube_india_data['trending_date']=pd.to_datetime(youtube_india_data['trending_date'],infer_datetime_format=True)
```

```python
youtube_india_data['publishedAt']=pd.to_datetime(youtube_india_data['publishedAt'],infer_datetime_format=True)
```

# Removing Null Data  

```python
youtube_india_data.dropna(subset=['channelTitle'],inplace=True)
```

# Filtering of Data

```python
youtube_india_data_v1=youtube_india_data.set_index('publishedAt')
```

```python
youtube_last_yr=youtube_india_data_v1.last('365D')
```

```python
youtube_last_yr.sort_values(by='publishedAt')
```

```python
youtube_last_yr=youtube_last_yr.reset_index()
```

```python
print('Data from',min(youtube_last_yr['publishedAt']),'to',max(youtube_last_yr['publishedAt']))
```

```python
youtube_last_yr.info()
```

# Extracting Date

```python
youtube_last_yr['day']=youtube_last_yr['trending_date'].dt.day
```

```python
youtube_last_yr['month']=youtube_last_yr['trending_date'].dt.month
youtube_last_yr['year']=youtube_last_yr['trending_date'].dt.year
```

```python
youtube_last_yr.head(10)
```

# Category description

```python
youtube_cat_in=pd.read_json('IN_category_id.json')
```

# Parsing category id

```python
id=[]
cat=[]
for t in youtube_cat_in['items']:
    for x,y in t.items():
        if x=='id':
            id.append(y)
        if x=='snippet':
            for a,b in y.items():
                if a=='title':
                    cat.append(b)
                    break
cat_dict={
    'id':id,
    'category':cat
}
category_id=pd.DataFrame(cat_dict)
```

```python
category_id['id']=category_id['id'].astype('int64')
```

```python
category_id
```

# Analysis

# Which type of videos were trending the most in the last year

```python
trend_yr=youtube_last_yr[['video_id','categoryId','month']]
video_trend_yr=pd.DataFrame(trend_yr.groupby(by='categoryId').video_id.nunique().sort_values(ascending=False)).merge(category_id,how='inner',left_on='categoryId',right_on='id').rename(columns={'video_id':'video_count'})
```

```python
sns.barplot(x=video_trend_yr['video_count'],y=video_trend_yr['category'],palette='turbo')
```

# Which Channel has their video trending the most 

```python
top10_channel=pd.DataFrame(youtube_last_yr.groupby(by='channelTitle').video_id.nunique().sort_values(ascending=False)).rename(columns={'video_id':'video_count'}).head(10)
```

```python
sns.color_palette("hls",8)
sns.barplot(x=top10_channel['video_count'],y=top10_channel.index,palette='turbo')
```

# Top 3 Channel whose videos end up on trending by eachmonth

```python
chan_trend_month=pd.DataFrame(youtube_last_yr.groupby(['month','channelTitle']).video_id.nunique().sort_values(ascending=False)).rename(columns={'video_id':'video_count'}).reset_index()
```

```python
for m in range(1,13):
    print(chan_trend_month.loc[chan_trend_month.month==m].head(3))
```

# which type of  videos take the shortest time to get on trending page after publishing

```python
df1=youtube_last_yr[['video_id','publishedAt','channelTitle','categoryId']]
df1=df1[~df1.duplicated()]
df2=youtube_last_yr['trending_date']
df1=df1.join(df2)
df1=df1.join([df1['trending_date'].dt.date-df1['publishedAt'].dt.date])
```

```python
df1=df1.rename(columns={0:'deltatrend'})
cat_deltatrend=pd.DataFrame(df1.groupby(by='categoryId').deltatrend.mean().sort_values(ascending=False)).merge(category_id,how='left',left_on='categoryId',right_on='id').dropna()
```

```python
sns.barplot(x=cat_deltatrend['deltatrend']/np.timedelta64(1,'D'),y=cat_deltatrend['category'],palette='magma')
```

# Which Channel's videos take the shortest time to get on trending page after publishing

```python
df1.groupby(by='channelTitle').deltatrend.mean().sort_values(ascending=True).head(10)
```

# which type of video remained longest on the trending page ?

```python
df11=youtube_last_yr.groupby(['video_id','channelTitle','categoryId']).apply(lambda x:max(x.trending_date)-min(x.trending_date))
df11=pd.DataFrame(df11).rename(columns={0:'trending_days'})
cat_remain_trend=pd.DataFrame(df11.groupby(['categoryId']).trending_days.mean().sort_values(ascending=False)).merge(category_id,how='left',left_on='categoryId',right_on='id')
```

```python
sns.barplot(x=cat_remain_trend['trending_days']/np.timedelta64(1,'D'),y=cat_remain_trend['category'],palette='plasma')
```

# Which channel's video remained the longest on the trending page 

```python
chan_remain_trend=pd.DataFrame(df11.groupby(['channelTitle']).trending_days.mean().sort_values(ascending=False)).head(15)
```

```python
sns.barplot(x=chan_remain_trend['trending_days']/np.timedelta64(1,'D'),y=chan_remain_trend.index,palette='rocket')
```
